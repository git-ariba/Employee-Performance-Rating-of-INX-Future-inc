### Inter Quartile Range

Note: 7 Columns are found with outliers. column Name:columns with outliers: NumCompaniesWorked, TrainingTimesLastYear, WorkExp_Total, WorkExp_Company,CurrentRoleExp,YearsWithCurrManager, YearsSinceLastPromotion 

```python
from scipy import stats
#importing IQR frim scipy, stats library
```

```python
#1.NumCompaniesWorked -to check if there are less than 1.5% outliers and impute them with median

iqr = stats.iqr(d['NumCompaniesWorked'],interpolation='midpoint')
print("IQR:",iqr)

# Getting first & third quantile
Q1 = d['NumCompaniesWorked'].quantile(0.25)
Q3 = d['NumCompaniesWorked'].quantile(0.75)

# Get maximum and minimum limit
min_limit = Q1 - 1.5*iqr
print("Minimum limit:",min_limit)

max_limit = Q3 + 1.5*iqr
print("Maximum Limit:",max_limit)

outliers = d[(d['NumCompaniesWorked'] > max_limit )]
 # | (d['NumCompaniesWorked'] < min_limit)] for the lowerlimit 
outlier_percentage = (len(outliers) / len(d)) * 100

print(f"Percentage of outliers before imputing: {outlier_percentage:.2f}%")
```

```python
#2.TrainingTimesLastYear- -to check if there are less than 1.5% outliers and impute them with median

iqr = stats.iqr(d['TrainingTimesLastYear'],interpolation='midpoint')
print("IQR:",iqr)

# Getting first & third quantile
Q1 = d['TrainingTimesLastYear'].quantile(0.25)
Q3 = d['TrainingTimesLastYear'].quantile(0.75)

# Get maximum and minimum limit
min_limit = Q1 - 1.5*iqr
print("Minimum limit:",min_limit)

max_limit = Q3 + 1.5*iqr
print("Maximum Limit:",max_limit)

outliers = d[(d['TrainingTimesLastYear'] > max_limit )]|(d['TrainingTimesLastYear'] < min_limit) #for the lowerlimit 
outlier_percentage = (len(outliers) / len(d)) * 100

print(f"Percentage of outliers before imputing: {outlier_percentage:.2f}%")
```

# 4.1 Random Forest -HyperParameter Tuning (GridSearchCV) 

**Parameters**
- **min_samples_split**: specifies the minimum number of samples required to split an internal node in the decision tree. setting it to a higher value can make the tree simpler and more robust against overfitting, hence hyperparameter tuning is done to select the best parameter.

- **criterion**:determines the function used to measure the quality of a split. it measures the impurity In classification problems, 'gini' and 'entropy', are commonly used criterias.

- **min_samples_leaf**: This parameter specifies the minimum number of samples required to be at a leaf node. If, after a split, any child node has fewer samples than this value, further splitting is stopped, and it becomes a leaf node. Similar to min_samples_split, setting it to higher values can simplify the tree and reduce overfitting.
